## Importing Data

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("Cleaned_Data.csv")
```

```python
df.shape
```

```python
df = df.iloc[:,1:]
```

```python
df.head()
```

```python
df["NoOfBathrooms"] = df["NoOfBathrooms"].apply(int)
df["Price"] = df["Price"].apply(int)
```

```python
df["Price"].max()
```

## Plotting Graphs

```python
plt.scatter(df["NoOfBathrooms"],df["Price"])
plt.xlabel("No of Bathrooms")
plt.ylabel("Price")
```

```python
plt.scatter(df["NoOfBedrooms"],df["Price"])
plt.xlabel("No of Bedrooms")
plt.ylabel("Price")
```

```python
plt.scatter(df["AreaSqYards"],df["Price"])
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
plt.scatter(df["Address"],df["Price"])
plt.xlabel("Area")
plt.ylabel("Price")
```
