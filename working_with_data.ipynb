```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("karachi_house_prices.csv")
```

```python
df.drop(columns=["Unnamed: 0"],inplace=True)
df
```

```python
# convert column to numeric data type, replacing invalid values with NaN
arr = pd.to_numeric(df["NoOfBathrooms"], errors="coerce")
```

```python
df["NoOfBathrooms"] = arr
```

```python
df["NoOfBedrooms"] = pd.to_numeric(df["NoOfBedrooms"], errors="coerce")
```

```python
df.sample(15)
```

```python
df.head()
```

```python
import re

df["AreaSqYards"] = df["AreaSqYards"].str.extract(r'(\d+(?:\.\d+)?)').astype(int)
```

```python
df.head()
```

```python
df["Price"].unique
```

```python
# extract numeric value using regular expressions
def func(x):
    return float(re.findall("\d+\.*\d*", x)[0])

df.loc[df['Price'].astype(str).str.contains('Lakh'), 'Price'] = df.loc[df['Price'].astype(str).str.contains('Lakh'), 'Price'].apply(func) * 10**5
df.loc[df['Price'].astype(str).str.contains('Crore'), 'Price'] = df.loc[df['Price'].astype(str).str.contains('Crore'), 'Price'].apply(func) * 10**7

```

```python
df.to_csv("Karachi_House_Price_Dataset.csv")
```
